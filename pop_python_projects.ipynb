```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('popular_python_projects.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
#SCATTER PLOT BETWEEN STARS AND FORKS OF THE PYTHON PROJECTS

plt.scatter(df['stars'], df['forks'])

#add axis labels
plt.xlabel('Stars')
plt.ylabel('Forks')
```

```python
#SCATTER PLOT BETWEEN STARS AND ISSUES OF THE PYTHON PROJECTS

plt.scatter(df['stars'], df['issues'])

#add axis labels
plt.xlabel('Stars')
plt.ylabel('Issues')
```

```python
#Conclusion:
'''
>Projects with more number of issues are having less number of stars
>Projects having lower number of stars have higher number of forks
'''
```
